# Matplotlib

Matplotlib is a low level graph plotting library in python that serves as a visualization utility

```python
import matplotlib
print(matplotlib.__version__)
```

# Pyplot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y)
plt.show()
```

To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8, 5])
ypoints = np.array([3, 10, 7])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

# Create Labels for a Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(['India', 'Russia', 'Brazil', 'USA', 'Shrilanka', 'China'])
y = np.array([240, 200, 260, 70, 20, 320])

plt.plot(x, y)

plt.xlabel("Countries")
plt.ylabel("Population(millions)")

plt.show()
```

# Title

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(['India', 'Russia', 'Brazil', 'USA', 'Shrilanka', 'China'])
y = np.array([240, 200, 260, 70, 20, 320])

plt.plot(x, y)
plt.title("Population Growth")
plt.xlabel("Countries")
plt.ylabel("Population(millions)")

plt.show()
```

# Set Font Properties for Title and Labels

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Population Growth of India per year", fontdict = font1)
plt.xlabel("Year", fontdict = font2)
plt.ylabel("Population(millions)", fontdict = font2)

plt.plot(x, y)
plt.show()
```

# Display Multiple Plots

```python
import matplotlib.pyplot as plt
import numpy as np

# plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) # the figure has 1 row, 2 columns, and this plot is the first plot.

plt.plot(x,y)

# plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) # the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x,y)

plt.show()
```

```python

# plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 1, 1)
plt.title('One',loc='left')
plt.plot(x,y)

# plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(3, 1, 2)
plt.title('Two',loc='left')
plt.plot(x,y)

# plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(3, 1, 3)
plt.title('Three',loc='left')
plt.plot(x,y)

plt.show()
```

# Scatter Plots

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['India', 'Russia', 'Brazil', 'USA', 'Shrilanka', 'China'])
y = np.array([240, 200, 260, 70, 20, 320])

plt.scatter(x, y)
plt.xlabel('Country')
plt.ylabel('Population(million)')
plt.title("Population",loc='left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# dead
x = np.array(['India', 'Russia', 'Brazil', 'USA', 'Shrilanka', 'China'])
y = np.array([40, 20, 60.2, 15.6, 5.7, 90.8])
plt.scatter(x, y,c='r')

# recovered
x = np.array(['India', 'Russia', 'Brazil', 'USA', 'Shrilanka', 'China'])
y = np.array([50, 13, 24.2, 8.6, 4.7, 20.8])
plt.scatter(x, y,c='g')

plt.xlabel('Country')
plt.ylabel('Population(Thousands)')
plt.title("Covid-19 Recovery & Death Graph Worldwide",loc='left')
plt.show()
```

# Color Each Dot

# Alpha

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# Start Angle

```python
y = np.array([35, 25, 25, 15])
mylabels = ["India", "China", "Brazil", "USA"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

# Explode

```python
y = np.array([35, 25, 25, 15])
mylabels = ["India", "China", "Brazil", "USA"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

# Shadow

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["red", "yellow", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

# Legend

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

# Legend With Header

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
